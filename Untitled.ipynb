<span style="font-size:30px;">**1** Load Dataset</span>

```python
import numpy as np
import pandas as pd
from numpy import loadtxt
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
dataset = pd.read_csv("/Users/albertegi/Dev/Diabetes_Prediction_Using_ANN/sample_data/diabetes.csv")
print(dataset.head())
```

```python
X = dataset.iloc[:, 0:8]  # Select all rows and columns 0 to 7
y = dataset.iloc[:, 8]    # Select column 8 (target variable)
```

```python
X
```

<span style="font-size:30px;">**2** Define the Model</span>
### This Neural Network is a fully connected network with three layers
1. The model expects rows of data with 8 variables (the input_dim = 8 argument)
2. The first hidden layer has 12 nodes and uses the relu activation function
3. The second hidden layer 9 nodes and uses the relu activation function
4. The output layer has one node and uses the sigmoid activation function.

 **Note** That the shape of the input to the model is defined as an argument on the first hidden layer
This means that the line of code that adds the first Dense layer does two things. Defining the input and the first hidden layer

```python
model = Sequential()
model.add(Dense(units=12, input_dim=8, activation='relu')) # input dimension and first layer
model.add(Dense(units=9, activation='relu')) # second layer
model.add(Dense(units=1,activation='sigmoid')) # third layer also called the output layer

# model = Sequential([
#     Input(shape=(8,)),  # Define input shape explicitly
#     Dense(units=12, activation='relu'),  # First hidden layer
#     Dense(units=9, activation='relu'),   # Second hidden layer
#     Dense(units=1, activation='sigmoid') # Output layer
# ])
```

<span style="font-size:30px;">**3** Compile the Model</span> 

Now that the model is defined, we can compile it (i.e. we need to put all our recipe
that we need together in order to get ready for training).

We must specify the loss function to use to evaluate a set of weights, the optimizer is is used to search
through different weights for the network and any optional metrics we would like to collect and report during training.
in this case, we will use cross entropy as the loss argument. this loss is for a binary classification problems and is defined in keras as **binary_crossentropy.**

We will define the optimizer as the efficient stochastic gradient descent algorithm **"adam".** This is a popular version of of gradient descent because it automatically tunes itself and gives good result in a wide range of problems.

Finally, because it is a classification problem, we will collect and report the classification accuracy via the mtrics argument.

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

<span style="font-size:30px;">**4** Fit the Model</span> 

We have defined our model and compiled it ready for efficient training.
Now it is time to execute the model on some data. We can train or fit our model on our loaded data by calling the **fit** function on the model.

The training occurs over epochs and each epoch is split into batches.
1. **Epoch:** one pass through all of the rows in the training dataset
2. **Batch:** One or more samples considered by the model within an epoch before weights are updated.
For this problem, we will run for a small number of epochs (150) and use a relatively small batch size of 10
                                               

```python
model.fit(X, y, epochs=150, batch_size=10)
```

We can see that we have 768 records, so out of that, we choose 10 records at a time (referred to as batch size) for 1 iteration (referred to as one epoch)
